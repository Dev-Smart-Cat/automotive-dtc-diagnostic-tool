```python
# Create the connection with the database from the env vars
conn = pg.connect(
    host=os.getenv("HOST_NAME"),
    port=os.getenv("PORT_NUMBER"),
    dbname=os.getenv("DB_NAME"),
    user=os.getenv("USER_NAME"),
    password=os.getenv("PASSWORD")
)
```

```python
# psycopg2 object responsible for execute queries
cur = conn.cursor()
```

```python
# Iterate over the list with the file names
for file_name in file_names_list:
    
    # Join the file path to file_name
    file_path = os.path.join(csv_folder, file_name)

    # Read the .csv file and save as a dataframe to remove duplicates
    df = pd.read_csv(file_path)
    # Remove duplicated dtc on column "code"
    df.drop_duplicates(subset=["code"], keep="first").to_csv(file_path, index=False)
```

```python
# Iterate over the list with the file names
for file_name in file_names_list:
    # Remove the placeholder .csv from file name
    table_name = file_name.replace(".csv", "")
    # Join the file path to file_name
    file_path = os.path.join(csv_folder, file_name)

    # Open the file as f to be copied to the postgres
    with open(file_path, "r", encoding="utf-8") as f:
        cur.execute(f"TRUNCATE TABLE {table_name} RESTART IDENTITY;")
        # Connect to the db through cur and
        # copy_expert: used to hanle SQL copy statement to copy outside files to Postgres inside tables 
        # and vice-versa
        cur.copy_expert(
            # SQL command to copy outside files to Postgres inside table
            # table_name: table created in the Postgres
            f"COPY {table_name} FROM STDIN WITH CSV HEADER",
            f
        )

    # Execute the copy with commit() method and print a message to confirm the process
    conn.commit()
    print(f"{table_name} imported.")

# VERY IMPORTAT, ALWAYS close the database connection
cur.close()
conn.close()
```
